````python
import pandas as pd
df=pd.read_csv('/content/diabetes.csv')
````

````python
df.head()
````

````python
import numpy as np
import torch
import torch.nn as nn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
````

````python
X_numpy=np.array(df.drop(['Outcome'],axis=1))
y_numpy=np.array(df['Outcome'])
y_numpy=y_numpy.reshape(-1,1)
n_samples, n_features = X_numpy.shape
X = torch.from_numpy(X_numpy.astype(np.float32))      #torch.from_numpy(): Converts a NumPy array into a PyTorch tensor.
y = torch.from_numpy(y_numpy.astype(np.float32))      #.astype(np.float32): Ensures that the NumPy arrays are of type float32, which is the default for PyTorch tensors.
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
````

````python
class Model(nn.Module):
    def __init__(self, n_input_features):
        super(Model, self).__init__()
        self.l1 = nn.Linear(n_input_features, 20)
        self.l2=nn.Linear(20,1)


    def forward(self, x):
        x = torch.relu(self.l1(x))
        x = torch.sigmoid(self.l2(x))
        return x
````

````python
model = Model(n_features)
````

````python
num_epochs = 1000
learning_rate = 0.001
criterion = nn.BCELoss() #Binary cross entropy
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
````

**a. Forward Pass (outputs = model(X_train))**
 -- The model uses the current weights to compute predictions (outputs).

**b. Loss Calculation (loss = criterion(outputs, y_train))** --
The loss is calculated to measure how far off the predictions are from the true values.

**c. Backward Pass (loss.backward())** --
Gradients of the loss with respect to the weights are computed via backpropagation.
These gradients tell the optimizer how much and in which direction to change each weight.

**d. Clear Gradients (optimizer.zero_grad())** --
This resets the gradients from the previous step so that new gradients can be calculated for the current iteration.
It does not erase or affect the weights themselves.

**e. Optimization Step (optimizer.step())** --
The optimizer updates the weights using the current gradients:
Example (simplified formula for gradient descent):

New weight = old weight - Î·*Graident

````python
for epoch in range(num_epochs):
    y_pred = model(X_train)
    loss = criterion(y_pred, y_train)

    loss.backward()
    optimizer.step()

    optimizer.zero_grad()

    if (epoch+1) % 10 == 0:
        print(f'epoch: {epoch+1}, loss = {loss.item():.4f}')
````

````python
model.eval()
y_pred = model(X_test)
test_loss = criterion(y_pred, y_test)
print('test loss is {}'.format(test_loss))
````
